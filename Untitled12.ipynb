```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8]
y=[4,6,7,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,marker ='o')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```
